## Task 1: Simple Calculator

```python
print("Simple Calculator")
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Addition:", num1 + num2)
print("Subtraction:", num1 - num2)
print("Multiplication:", num1 * num2)

if num2 != 0:
    print("Division:", num1 / num2)
else:
    print("Division: Cannot divide by zero")

print("Power:", num1 ** num2)
```

## Task 2: Check if list contains numeric value

```python
my_list = ["Hizar", "Ali", 25, "AI", 3.14]

found = False

for item in my_list:
    if isinstance(item, (int, float)):
        found = True
        break

if found:
    print("The list contains numeric value.")
else:
    print("No numeric value found.")
```

## Task 4: Sum numeric values in dictionary

```python
marks = {
    "Math": 80,
    "English": 75,
    "Science": 85
}

total = 0

for value in marks.values():
    total += value

print("Total Marks:", total)
```

## Task 5: Identify duplicate values in list

```python
numbers = [1, 2, 3, 4, 2, 5, 6, 3, 7]

duplicates = []

for num in numbers:
    if numbers.count(num) > 1 and num not in duplicates:
        duplicates.append(num)

print("Duplicate values:", duplicates)
```
